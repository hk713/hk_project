## #부산행

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

pres=['https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=130966&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=highest&page=',
     'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=130966&type=after&onlyActualPointYn=N&onlySpoilerPointYn=N&order=lowest&page=']

review = []
rate = []
target = []

for pre in pres:
    for i in range(1,200):
        url = pre+str(i)
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'html.parser')
        
        id_list = []
        id_pre = '_filtered_ment_'
        
        for i in range(10):
            id_list.append(id_pre+str(i))
        
        for id in id_list:
            review.append(soup.find('span',{'id':id}).get_text().strip())
            
        rate_list = []
        rate_list = (soup.select('div.star_score > em'))
        
        for r in rate_list:
            r = int(re.sub('<.+?>','',str(r)))
            rate.append(r)
            if r>=8:
                target.append('good')
            elif r<=4:
                target.append('bad')
            else:
                target.append('soso')

df = pd.DataFrame({'review':review, 'rate':rate, 'target':target})
```

## 트레이닝/테스트셋 분리

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(review, target, test_size=0.2, random_state=0)
```

```python
len(train_x), len(train_y)
```

```python
len(test_x), len(test_y)
```

## 특징추출

```python
from konlpy.tag import Twitter
from sklearn.feature_extraction.text import TfidfVectorizer
twitter = Twitter()

tfv = TfidfVectorizer(tokenizer=twitter.morphs, ngram_range=(1,2), min_df=3, max_df=0.9)
tfv.fit(train_x)

tfv_train_x = tfv.transform(train_x)
tfv_train_x
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

clf = LogisticRegression(random_state=0)
params = {'C':[1,3,5,7,9]}
grid_cv = GridSearchCV(clf, param_grid=params, cv=4, scoring='accuracy', verbose=1)
grid_cv.fit(tfv_train_x, train_y)
```

```python
tfv_test_x = tfv.transform(test_x)
grid_cv.best_estimator_.score(tfv_test_x, test_y)
```
