# 과제#1
네이버 영화 사이트에서 자신이 원하는 영화를 골라서 평점과 리뷰내용을 1페이지에서 10페이지까지 100개 크롤링 하시오.

```python
import requests
from bs4 import BeautifulSoup

# 영화<비긴어게인>
site_pre ="https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=96379&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="

id_list=[]
id_pre="_filtered_ment_"
for i in range(10):
    id_list.append(id_pre+str(i))

for i in range(10):
    site=site_pre+str(i+1)
    res=requests.get(site)
    
    soup = BeautifulSoup(res.content,'html.parser')
    
    # 평점
    score_list=[]
    scores = soup.find_all('div','star_score')
    for score in scores:
        score_list.append(score.get_text())
    
    # 리뷰내용
    mydata=[]
    for id in id_list:
        mydata.append(soup.find('span',{'id':id}).get_text().strip())
    
    for score, review in zip(score_list,mydata):
        print(score.strip(),review)         
```

# 과제#2
같은 영화를 다음 영화에서 검색하여 동일한 형태로 크롤링 하시오.

```python
import requests
from bs4 import BeautifulSoup

site_pre="https://movie.daum.net/moviedb/grade?movieId=80780&type=netizen&page="

for i in range(10):
    site=site_pre+str(i+1)
    res=requests.get(site)
    
    soup = BeautifulSoup(res.content,'html.parser')
    
    # 평점
    score_list=[]
    scores = soup.find_all('em','emph_grade')
    for score in scores:
        score_list.append(score.get_text())
    
    # 리뷰내용
    mydata=[]
    reviews = soup.find_all('p','desc_review')
    for review in reviews:
        mydata.append(review.get_text("",strip=True))
    
    for score, review in zip(score_list,mydata):
        print(score.strip(),review)
```
