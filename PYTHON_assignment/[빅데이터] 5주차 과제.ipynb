과제#2 쇼핑몰 특정 페이지의 물품 리스트를 가져오기

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('http://www.homeplus.co.kr/app.exhibition.bestshop.BestShop.ghs?comm=usr.bestshop&WT.ac=Main_GNB_best')
```

```python
soup = BeautifulSoup(res.content,'html.parser')
```

```python
mydata = soup.find_all('a','name')
len(mydata)
```

```python
product=[]
```

```python
for item in mydata:
    product.append(item.get_text())
```

```python
len(product)
```

```python
mydata = soup.find_all('strong','buy')
len(mydata)
```

```python
price=[]
for item in mydata:
    price.append(item.get_text()+"원")
```

```python
len(price)
```

```python
import pandas as pd
```

```python
table = pd.DataFrame(columns=['물품','가격'])
```

```python
table['물품']=product
```

```python
table['가격']=price
```

```python
table
```

```python
#왼쪽 정렬
table=table.style.set_properties(**{'text-align': 'left'}).set_table_styles([ dict(selector='th', props=[('text-align', 'left')] ) ])
```
